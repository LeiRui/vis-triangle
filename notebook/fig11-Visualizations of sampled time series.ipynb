```python
from myfuncs import *
import glob
import re

pngDir="./examples" # dir of line chart results
if not os.path.exists(pngDir):
    os.makedirs(pngDir)

def get_first_matched_file(folder_path,pattern):
    pattern = re.compile(pattern)
    files = os.listdir(folder_path)
    matched_files = [file for file in files if pattern.match(file)]
    if matched_files:
        return os.path.join(folder_path,matched_files[0])
    else:
        return None

dataDir="../datasets"
d1=os.path.join(dataDir,"Qloss-small.csv")
d2=os.path.join(dataDir,"Pyra1-small.csv")
d3=os.path.join(dataDir,"WindSpeed-small.csv")
d4=os.path.join(dataDir,"RTD-small.csv")

headerList=[None,None,None,None]
titleList=[
    '(a) Qloss',
    '(b) Ryra1',
    '(c) WindSpeed',
    '(d) RTD'
]
dList=[d1,d2,d3,d4]
    
visvalAreaArray=[0.6205000877380371,1899451.5554213524,14701.5,76155.67999982834]
segmentDir="segmentResults"
simpieceResultFiles=[r"Qloss-2500000-\d+-simpiece.csv",
                r"Pyra1-2500000-\d+-simpiece.csv",
                r"WindSpeed-2500000-\d+-simpiece.csv",
                 r"RTD-2500000-\d+-simpiece.csv"]
fswResultFiles=[r"Qloss-2500000-\d+-fsw.csv",
                 r"Pyra1-2500000-\d+-fsw.csv",
                 r"WindSpeed-2500000-\d+-fsw.csv",
                 r"RTD-2500000-\d+-fsw.csv"]

tmp=simpieceResultFiles
for i in range(len(tmp)):
    first_matched_file = get_first_matched_file(segmentDir,tmp[i])
    tmp[i] = first_matched_file
    
tmp=fswResultFiles
for i in range(len(tmp)):
    first_matched_file = get_first_matched_file(segmentDir,tmp[i])
    tmp[i] = first_matched_file
    
anti=True
lw=1
width=250
height=150
dpi=144

for i in range(len(dList)):
    if i==2:
        nout=200
    else:
        nout=100
    
    filename = dList[i]
    df=pd.read_csv(filename,header=headerList[i]) # default no header
    v=df.iloc[:,1]
    v=v.to_numpy(dtype='float')
    t=np.arange(0,len(v))

    name=os.path.split(filename)[1].split(".")[0]+'_'+'original'
    downsample='original'
    subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,isPlot=False,pngDir=pngDir)

    ##############################################################
    name=os.path.split(filename)[1].split(".")[0]+'_'+'LTTBETDownsampler' 
    downsample='LTTBETDownsampler'
    t2,v2=subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,isPlot=False,pngDir=pngDir)
    print(len(t2))

    ##############################################################
    name=os.path.split(filename)[1].split(".")[0]+'_'+'LTTBETFurtherDownsampler' 
    downsample='LTTBETFurtherDownsampler'
    t3,v3=subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,isPlot=False,pngDir=pngDir)
    print(len(t3))

    ##############################################################
    name=os.path.split(filename)[1].split(".")[0]+'_'+'MinMaxLTTB2Downsampler'
    downsample='MinMaxLTTB2Downsampler'
    t4,v4,t4_pre,v4_pre=subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,\
                                               isPlot=False,pngDir=pngDir)
    print(len(t4))

    ##############################################################
    name=os.path.split(filename)[1].split(".")[0]+'_'+'MinMaxDownsampler'
    downsample='MinMaxDownsampler'
    t5,v5=subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,isPlot=False,pngDir=pngDir)
    print(len(t5))

    ##############################################################
    name=os.path.split(filename)[1].split(".")[0]+'_'+'M4Downsampler'
    downsample='M4Downsampler'
    t6,v6=subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,isPlot=False,pngDir=pngDir)
    print(len(t6))

    ##############################################################
    name=os.path.split(filename)[1].split(".")[0]+'_'+'fastVisval'
    downsample='fastVisval'
    area=visvalAreaArray[i]
    t8,v8=subplt_myplot_external(width,height,dpi,name,anti,downsample,area,lw,t,v,isPlot=False,pngDir=pngDir)
    print(len(t8))

    ##############################################################
    name=os.path.split(filename)[1].split(".")[0]+'_'+'uniformTime'
    downsample='uniformTime'
    t10,v10=subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,isPlot=False,pngDir=pngDir)
    print(len(t10))

    #############################################################
    resultFile=fswResultFiles[i]
    name=os.path.split(filename)[1].split(".")[0]+'_'+'FSW'
    downsample='FSW'
    t11,v11=subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,isPlot=False,\
                                   resultFile=resultFile,pngDir=pngDir)
    print(len(t11))

    ##############################################################
    resultFile=simpieceResultFiles[i]
    name=os.path.split(filename)[1].split(".")[0]+'_'+'SimPiece'
    downsample='SimPiece'
    t12,v12=subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,isPlot=False,\
                                   resultFile=resultFile,pngDir=pngDir)
    print(len(t12))
    
    ##############################################################
    resultFile=simpieceResultFiles[i]
    name=os.path.split(filename)[1].split(".")[0]+'_'+'DFT'
    downsample='PyDFT'
    t13,v13=subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,isPlot=False,pngDir=pngDir)
    print(len(t13)) # note already constructed for plotting
    
    ##############################################################
    resultFile=simpieceResultFiles[i]
    name=os.path.split(filename)[1].split(".")[0]+'_'+'PCA'
    downsample='PCA'
    t14,v14=subplt_myplot_external(width,height,dpi,name,anti,downsample,nout,lw,t,v,isPlot=False,pngDir=pngDir)
    print(len(t14)) # note already constructed for plotting

    ##############################################################
    # original line chart in green separately
    name=os.path.split(filename)[1].split(".")[0]+'_'+'original'+'_green'
    full_frame(width,height,dpi)
    v_min=min(v)
    v_max=max(v)
    t_min=min(t)
    t_max=max(t)
    plt.plot(t,v,'-',color='tab:green',linewidth=lw,antialiased=anti)
    plt.xlim(t_min, t_max)
    plt.ylim(v_min, v_max)
    plt.savefig(os.path.join(pngDir,name+'.png'),backend='agg')
    plt.close()
    
print('finish')
```
